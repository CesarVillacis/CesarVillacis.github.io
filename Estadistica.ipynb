1. Obtenga la media de un conjunto de números y su correspondiente gráfica

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos: conjunto de números
numeros = [10, 50, 20, 25, 30]
#calculo de la media
suma=0
for elemento in numeros:
  suma+=elemento

media=suma/len(numeros)

print(media)

# Graficar los números y la media
plt.plot(numeros, marker='o', label='Números')
plt.axhline(y=media, color='r', linestyle='--', label=f'Media: {media:.2f}')

# Configuraciones adicionales de la gráfica
plt.title('Conjunto de números y su media')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.legend()

# Mostrar la gráfica
plt.show()

```

2. Obtención de la mediana de dos conjuntos de datos (uno con número par e impar de elementos) y sus gráficas

```python
# Conjunto de datos pares e impares
conjunto_par = [10, 20, 30, 40]
conjunto_impar = [5, 16, 25, 35, 45]

# Calcular la mediana de ambos conjuntos
mediana_par = np.median(conjunto_par)
mediana_impar = np.median(conjunto_impar)

print(f"La mediana del conjunto par es: {mediana_par}")
print(f"La mediana del conjunto impar es: {mediana_impar}")

# Graficar ambos conjuntos con sus medianas
plt.figure(figsize=(10, 5))

# Gráfico del conjunto par
plt.subplot(1, 2, 1)
plt.plot(conjunto_par, marker='o', label='Conjunto Par')
plt.axhline(y=mediana_par, color='r', linestyle='--', label=f'Mediana: {mediana_par:.2f}')
plt.title('Conjunto Par y su Mediana')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.legend()

# Gráfico del conjunto impar
plt.subplot(1, 2, 2)
plt.plot(conjunto_impar, marker='o', label='Conjunto Impar')
plt.axhline(y=mediana_impar, color='r', linestyle='--', label=f'Mediana: {mediana_impar:.2f}')
plt.title('Conjunto Impar y su Mediana')
plt.xlabel('Índice')
plt.ylabel('Valor')
plt.legend()

# Mostrar ambas gráficas
plt.tight_layout()
plt.show()
```

3. Obtención de la moda de un conjunto de datos

```python
from scipy import stats
import matplotlib.pyplot as plt

# Conjunto de datos
conjunto_moda = [10, 20, 20, 30, 40, 40, 40, 50]

# Calcular la moda
moda = stats.mode(conjunto_moda, keepdims=True)  # Asegura que el resultado esté en formato de array
print(f"La moda es: {moda.mode[0]} con una frecuencia de {moda.count[0]}")

# Graficar el conjunto de datos y resaltar la moda
plt.hist(conjunto_moda, bins=range(min(conjunto_moda), max(conjunto_moda) + 10, 10), alpha=0.7, color='blue', rwidth=0.85)
plt.axvline(x=moda.mode[0], color='r', linestyle='--', label=f'Moda: {moda.mode[0]}')

# Configuraciones adicionales de la gráfica
plt.title('Distribución del conjunto de datos y su moda')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar la gráfica
plt.show()


```
